# Import

This notebook connects to the Mongo DB and dumps the Polar data to the trainingSession collection

```python
import pymongo
import json
from pymongo import InsertOne
from pathlib import Path

client = pymongo.MongoClient("mongodb://root:example@localhost:27017/")
db = client.polar
collection = db.trainingSession
requesting = []

root = Path("./polar-data")

for file in root.glob("training-session*"):
    with file.open() as f:
        training_data = json.load(f)
        requesting.append(InsertOne(training_data))

result = collection.bulk_write(requesting)
client.close()
```
